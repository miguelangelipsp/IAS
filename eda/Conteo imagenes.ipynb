### Contar archivos por hora, luego registrar el conteo por alimentador y posterior registrar fecha y hora.
_Nota_: Se realiza el conteo sobre las carpetas que empiezan por Alimentador/Fecha/Hora.

```python
import os
import csv

def generar_registro_csv_con_imgextraidas(imgextraidas_file, parent_dirs, csv_filename):
    """
    Lee el archivo 'imgextraidas.txt' (que contiene solo nombres de imágenes) y, 
    para cada carpeta de hora en cada alimentador y fecha, verifica si existe 
    alguna imagen cuyo nombre se encuentre en el archivo.
    
    Se escribe un CSV con las columnas:
      - codigoAlimentador: nombre de la carpeta base (código del alimentador).
      - Alimentador: 1 si existe al menos una imagen en esa carpeta de hora, 0 en caso contrario.
      - Fecha: el nombre de la carpeta de fecha.
      - Hora: el nombre de la carpeta de hora.
    """
    # Leer el archivo de imágenes extraídas y guardar en un set
    with open(imgextraidas_file, 'r', encoding='utf-8') as f:
        imagenes_extraidas = {line.strip() for line in f if line.strip()}

    rows = []
    
    # Recorrer cada directorio padre (alimentador)
    for parent in parent_dirs:
        if not os.path.isdir(parent):
            print(f"El directorio {parent} no existe. Se omitirá.")
            continue

        # El código del alimentador es el nombre de la carpeta base
        codigo_alimentador = os.path.basename(parent)
        
        # Listar los subdirectorios de fecha dentro del directorio padre
        fechas = [d for d in os.listdir(parent) if os.path.isdir(os.path.join(parent, d))]
        for fecha in fechas:
            path_fecha = os.path.join(parent, fecha)
            # Listar los subdirectorios de hora dentro de la carpeta de fecha
            horas = [h for h in os.listdir(path_fecha) if os.path.isdir(os.path.join(path_fecha, h))]
            for hora in horas:
                path_hora = os.path.join(path_fecha, hora)
                # Listar solo los nombres de los archivos en la carpeta de hora
                archivos = [f for f in os.listdir(path_hora) if os.path.isfile(os.path.join(path_hora, f))]
                # Verificar si existe al menos un archivo cuyo nombre esté en el set de imágenes extraídas
                existe_imagen = any(nombre in imagenes_extraidas for nombre in archivos)
                valor_alimentador = 1 if existe_imagen else 0
                # Se añade una fila con: [codigoAlimentador, Alimentador (1/0), Fecha, Hora]
                rows.append([codigo_alimentador, valor_alimentador, fecha, hora])
    
    # Escribir el archivo CSV con encabezado
    with open(csv_filename, 'w', newline='', encoding='utf-8') as csvfile:
        writer = csv.writer(csvfile)
        writer.writerow(["codigoAlimentador", "Alimentador", "Fecha", "Hora"])
        writer.writerows(rows)
    
    print(f"Se ha generado el archivo CSV '{csv_filename}' con {len(rows)} registros.")

if __name__ == "__main__":
    imgextraidas_file = r"C:\Users\miguel.castillo\Documents\PROYECTOS\AUTOMATIZACION_SOBRANTES\Dataset\MASK\imgextraidas_3.txt"
    
    # Lista de directorios padres (alimentadores)
    parent_dirs = [
        r"V:\Imagenes\U0529\2029",
        r"V:\Imagenes\U0529\2030",
        r"V:\Imagenes\U0529\2031",
        r"V:\Imagenes\U0529\2032",
        r"V:\Imagenes\U0529\2033",
        r"V:\Imagenes\U0529\2034",
        r"V:\Imagenes\U0529\2035",
        r"V:\Imagenes\U0529\2036",
        r"V:\Imagenes\U0529\2037",
        r"V:\Imagenes\U0529\2038",
        r"V:\Imagenes\U0529\2039",
        r"V:\Imagenes\U0529\2040",
        r"V:\Imagenes\U0529\2041",
        r"V:\Imagenes\U0529\2042"
    ]
    
    csv_filename = "registro_imagenes_alimentadores_3.csv"
    generar_registro_csv_con_imgextraidas(imgextraidas_file, parent_dirs, csv_filename)
```

### Contar archivos por fecha, hora y Alimentador

```python
import os
import csv

def generar_registro_csv_con_imgextraidas(imgextraidas_file, base_dirs, csv_filename):
    """
    Lee el archivo 'imgextraidas.txt' (que contiene solo nombres de imágenes) y,
    para cada estructura Fecha -> Hora -> Alimentador, verifica si en la carpeta de 
    Alimentador existe al menos una imagen cuyo nombre se encuentre en dicho archivo.
    
    Se escribe un CSV con las columnas:
      - codigoAlimentador: nombre de la carpeta de alimentador (última carpeta).
      - Alimentador: 1 si existe al menos una imagen en esa carpeta, 0 en caso contrario.
      - Fecha: el nombre de la carpeta de Fecha (directorio base).
      - Hora: el nombre de la carpeta de Hora.
    """
    # Leer el archivo de imágenes extraídas y guardar los nombres en un conjunto
    with open(imgextraidas_file, 'r', encoding='utf-8') as f:
        imagenes_extraidas = {line.strip() for line in f if line.strip()}

    rows = []

    # Recorrer cada directorio base (cada uno es una carpeta de Fecha)
    for base in base_dirs:
        if not os.path.isdir(base):
            print(f"El directorio {base} no existe. Se omitirá.")
            continue

        # La carpeta base es la Fecha
        fecha = os.path.basename(base)
        
        # Listar los directorios de Hora dentro de la carpeta Fecha
        horas = [h for h in os.listdir(base) if os.path.isdir(os.path.join(base, h))]
        for hora in horas:
            path_hora = os.path.join(base, hora)
            # Listar los directorios de Alimentador dentro de la carpeta Hora
            alimentadores = [a for a in os.listdir(path_hora) if os.path.isdir(os.path.join(path_hora, a))]
            for alimentador in alimentadores:
                path_alimentador = os.path.join(path_hora, alimentador)
                # Listar solo los nombres de los archivos en la carpeta de Alimentador
                archivos = [f for f in os.listdir(path_alimentador) if os.path.isfile(os.path.join(path_alimentador, f))]
                # Verificar si existe al menos un archivo cuyo nombre esté en el set de imágenes extraídas
                existe_imagen = any(nombre in imagenes_extraidas for nombre in archivos)
                valor_alimentador = 1 if existe_imagen else 0
                # Agregar la fila: [codigoAlimentador, Alimentador, Fecha, Hora]
                rows.append([alimentador, valor_alimentador, fecha, hora])
    
    # Escribir el CSV con encabezado
    with open(csv_filename, 'w', newline='', encoding='utf-8') as csvfile:
        writer = csv.writer(csvfile)
        writer.writerow(["codigoAlimentador", "Alimentador", "Fecha", "Hora"])
        writer.writerows(rows)
    
    print(f"Se ha generado el archivo CSV '{csv_filename}' con {len(rows)} registros.")

if __name__ == "__main__":
    imgextraidas_file = r"C:\Users\miguel.castillo\Documents\PROYECTOS\AUTOMATIZACION_SOBRANTES\Dataset\MASK\imgextraidas_abril_1.txt"
    
    # Lista de directorios base (cada uno es una carpeta de Fecha)
    base_dirs = [
        r"V:\Imagenes\U0529\2025-04-03",
        r"V:\Imagenes\U0529\2025-04-04",
        r"V:\Imagenes\U0529\2025-04-05",
        r"V:\Imagenes\U0529\2025-04-06",
        r"V:\Imagenes\U0529\2025-04-07",
        r"V:\Imagenes\U0529\2025-04-08",
        r"V:\Imagenes\U0529\2025-04-09",
        r"V:\Imagenes\U0529\2025-04-10",
        r"V:\Imagenes\U0529\2025-04-11",
        r"V:\Imagenes\U0529\2025-04-12",
        r"V:\Imagenes\U0529\2025-04-13",
        r"V:\Imagenes\U0529\2025-04-14"
    ]
    
    csv_filename = "registro_imagenes_alimentadores_abril_3a14.csv"
    generar_registro_csv_con_imgextraidas(imgextraidas_file, base_dirs, csv_filename)
```

```python
import pandas as pd
from datetime import datetime, time

def categorizar_turno(hora_str):
    """
    Recibe una cadena de hora y la convierte en un objeto time para clasificarla según:
      - Mañana: de 07:00 a 11:30
      - Tarde: de 12:00 a 18:00
      - Noche: de 18:10 a 06:00 (incluyendo las horas de madrugada)
      
    Se intenta primero con el formato "%H_%M_%S.%f" (por ejemplo, "17_54_00.092990")
    y, si falla, se intenta con "%H-%M" (por ejemplo, "09-30").
    En caso de no poder parsear la cadena se devuelve "sin clasificar".
    """
    try:
        # Intentar primero con formato que incluye segundos y microsegundos
        t = datetime.strptime(hora_str, "%H_%M_%S.%f").time()
    except ValueError:
        try:
            # Intentar con formato solo de horas y minutos separados por guion
            t = datetime.strptime(hora_str, "%H-%M").time()
        except ValueError:
            print(f"Error: formato de hora desconocido: {hora_str}")
            return "sin clasificar"
    
    if time(7, 0) < t < time(12, 0):
        return "mañana"
    elif time(12, 0) <= t <= time(18, 0):
        return "tarde"
    elif t >= time(18, 0) or t < time(7, 0):
        return "noche"
    else:
        return "sin clasificar"

def analizar_imagenes_multiple(csv_files):
    """
    Lee una lista de archivos CSV que contienen las columnas:
      - codigoAlimentador, Foto, Fecha y Hora.
    Une los datos y realiza los siguientes análisis:
      1. Por codigoAlimentador: cuántas imágenes buenas y malas existen.
      2. Por Fecha: cuántas imágenes buenas y malas se generaron por fecha.
      3. Por Turno: clasifica la columna Hora y cuenta cuántas imágenes buenas y malas se generaron en la mañana, tarde y noche.
    
    Retorna tres DataFrames con los resultados.
    """
    # Leer cada CSV y concatenarlos
    df_list = []
    for csv_file in csv_files:
        df = pd.read_csv(csv_file, encoding="utf-8")
        df_list.append(df)
    df_all = pd.concat(df_list, ignore_index=True)
    
    # Asegurarse de que la columna 'Foto' sea numérica (1 = buena, 0 = mala)
    df_all["Foto"] = pd.to_numeric(df_all["Foto"], errors="coerce")
    
    # --- Análisis por codigoAlimentador ---
    agrupado_alim = df_all.groupby("codigoAlimentador")["Foto"].agg(["sum", "count"]).reset_index()
    agrupado_alim = agrupado_alim.rename(columns={"sum": "Buenas", "count": "Total"})
    agrupado_alim["Malas"] = agrupado_alim["Total"] - agrupado_alim["Buenas"]
    
    # --- Análisis por Fecha ---
    agrupado_fecha = df_all.groupby("Fecha")["Foto"].agg(["sum", "count"]).reset_index()
    agrupado_fecha = agrupado_fecha.rename(columns={"sum": "Buenas", "count": "Total"})
    agrupado_fecha["Malas"] = agrupado_fecha["Total"] - agrupado_fecha["Buenas"]
    
    # --- Análisis por Turno (mañana, tarde, noche) ---
    # Se clasifica la columna "Hora" aplicando la función categorizar_turno.
    df_all["Turno"] = df_all["Hora"].apply(categorizar_turno)
    agrupado_turno = df_all.groupby("Turno")["Foto"].agg(["sum", "count"]).reset_index()
    agrupado_turno = agrupado_turno.rename(columns={"sum": "Buenas", "count": "Total"})
    agrupado_turno["Malas"] = agrupado_turno["Total"] - agrupado_turno["Buenas"]
    
    return agrupado_alim, agrupado_fecha, agrupado_turno

if __name__ == "__main__":
    # Lista de archivos CSV a analizar
    csv_files = [
        "registro_imagenes_alimentadores_1.csv",
        "registro_imagenes_alimentadores_2.csv",
        "registro_imagenes_alimentadores_3.csv",
        "registro_imagenes_alimentadores_1a4.csv",
        "registro_imagenes_alimentadores_2a5.csv",
        "registro_imagenes_alimentadores_3a6.csv"
    ]
    
    # Se obtienen los DataFrames con los resultados de cada análisis.
    agrupado_alim, agrupado_fecha, agrupado_turno = analizar_imagenes_multiple(csv_files)
    
    # Impresión de resultados:
    print("Análisis por codigoAlimentador (imágenes buenas y malas):")
    print(agrupado_alim)
    
    print("\nAnálisis por Fecha (imágenes buenas y malas):")
    print(agrupado_fecha)
    
    print("\nAnálisis por Turno (mañana, tarde, noche):")
    print(agrupado_turno)
```
